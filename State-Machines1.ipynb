ตัวอย่างที่ 1 เราสามารถกำหนด Finite-State Machine (FSM) ที่มี 2 สถานะคือ นอนหลับและตื่นนอน หากต้องการย้ายระหว่างสถานะ จะกำหนดช่วงการเปลี่ยน wakeup และ sleep
image.png

```python
from fysom import Fysom
fsm = Fysom({'initial' : 'awake', 'events' : [
    {'name': 'wakeup','src' : 'sleeping','dst':'awake'},
    {'name': 'sleep', 'scr' : 'awake','dst' : 'sleeping' }]})

print(fsm.current)
fsm.sleep()
print(fsm.current)
fsm.wakeup()
print(fsm.current)
```

```python
from fysom import Fysom
finit_state = Fysom({ 'initial': 's0',
    'events': [
      {'name': '1', 'src': 's0', 'dst': 's1'},
      {'name': '1', 'src': 's1', 'dst': 's2'},
      {'name': '1', 'src': 's2', 'dst': 's3'},
      {'name': '1', 'src': 's3', 'dst': 's3'},
      {'name': '5', 'src': 's0', 'dst': 's3'},
      {'name': '5', 'src': 's1', 'dst': 's3'},
      {'name': '5', 'src': 's2', 'dst': 's3'},
      {'name': '5', 'src': 's3', 'dst': 's3'},
      {'name': 'green_btn', 'src': 's0', 'dst': 's0'},
      {'name': 'green_btn', 'src': 's1', 'dst': 's1'},
      {'name': 'green_btn', 'src': 's2', 'dst': 's2'},
      {'name': 'green_btn', 'src': 's3', 'dst': 's0'},
      {'name': 'red_btn', 'src': 's0', 'dst': 's0'},
      {'name': 'red_btn', 'src': 's1', 'dst': 's1'},
      {'name': 'red_btn', 'src': 's2', 'dst': 's2'},
      {'name': 'red_btn', 'src': 's3', 'dst': 's0'} ] })

while True:
    name = input('Enter your input : ')
    str_name = str(name)
    if str_name == '1' and finit_state.isstate('s0'):
      print('you pay 1 baht')
      finit_state.trigger(str_name)
    elif str_name == '1' and finit_state.isstate('s1'):
      print('you pay 2 baht')
      finit_state.trigger(str_name)
    elif str_name == '1' and finit_state.isstate('s2'):
      print('you pay 3 baht and you can select green or red water ')
      finit_state.trigger(str_name)
    elif str_name == '1' and finit_state.isstate('s3'):
      print('you pay 4 baht and change to you 1 baht then you can select green or red water ')
      finit_state.trigger(str_name)
    elif str_name == '5' and finit_state.isstate('s0'):
      print('you pay 5 baht and change to you 2 baht then you can select green or red water ')
      finit_state.trigger(str_name)
    elif str_name == '5' and finit_state.isstate('s1'):
      print('you pay 6 baht and change to you 3 baht then you can select green or red water ')
      finit_state.trigger(str_name)
    elif str_name == '5' and finit_state.isstate('s2'):
      print('you pay 7 baht and change to you 4 baht then you can select green or red water ')
      finit_state.trigger(str_name)
    elif str_name == '5' and finit_state.isstate('s3'):
      print('you pay 8 baht and change to you 5 baht then you can select green or red water ')
      finit_state.trigger(str_name)
    elif str_name == 'green_btn' and finit_state.isstate('s0'):
      print('You do not have enough money')
      finit_state.trigger(str_name)
    elif str_name == 'green_btn' and finit_state.isstate('s1'):
      print('You do not have enough money')
      finit_state.trigger(str_name)
    elif str_name == 'green_btn' and finit_state.isstate('s2'):
      print('You do not have enough money')
      finit_state.trigger(str_name)
    elif str_name == 'green_btn' and finit_state.isstate('s3'):
      print('Thank you for the purchase.')
      finit_state.trigger(str_name)
      break
    elif str_name == 'red_btn' and finit_state.isstate('s0'):
      print('You do not have enough money')
      finit_state.trigger(str_name)
    elif str_name == 'red_btn' and finit_state.isstate('s1'):
      print('You do not have enough money')
      finit_state.trigger(str_name)
    elif str_name == 'red_btn' and finit_state.isstate('s2'):
      print('You do not have enough money')
      finit_state.trigger(str_name)
    elif str_name == 'red_btn' and finit_state.isstate('s3'):
      print('Thanks for the purchase.')
      finit_state.trigger(str_name)
      break
    else:
      print('ERROR')
      break
```

```python
from fysom import Fysom
finit_state = Fysom({ 'initial' : 's0', 'events' : [
      {'name': '0', 'src': 's0', 'dst': 's1'},
      {'name': '0', 'src': 's1', 'dst': 's1'},
      {'name': '0', 'src': 's2', 'dst': 's3'},
      {'name': '0', 'src': 's3', 'dst': 's1'},
      {'name': '0', 'src': 's4', 'dst': 's3'},
      {'name': '1', 'src': 's0', 'dst': 's3'},
      {'name': '1', 'src': 's1', 'dst': 's2'},
      {'name': '1', 'src': 's2', 'dst': 's4'},
      {'name': '1', 'src': 's3', 'dst': 's0'},
      {'name': '1', 'src': 's4', 'dst': 's4'} ]})
number = int(input('number of Input ='))
print('Start state',finit_state.current)
for i in range(number) :
  name = input('Enter your input :')
  str_name = str(name)
  if str_name == '0' and finit_state.isstate('s0'):
    finit_state.trigger (str_name)
    print('State',finit_state.current)
  elif str_name == '0' and finit_state.isstate('s1'):
    finit_state.trigger (str_name)
    print('State',finit_state.current)
  elif str_name == '0' and finit_state.isstate('s2'):
    finit_state.trigger (str_name)
    print('State',finit_state.current)
  elif str_name == '0' and finit_state.isstate('s3'):
    finit_state.trigger (str_name)
    print('State',finit_state.current)
  elif str_name == '0' and finit_state.isstate('s4'):
    finit_state.trigger (str_name)
    print('State',finit_state.current)
  elif str_name == '1' and finit_state.isstate('s0'):
    finit_state.trigger (str_name)
    print('State',finit_state.current)
  elif str_name == '1' and finit_state.isstate('s1'):
    finit_state.trigger (str_name)
    print('State',finit_state.current)
  elif str_name == '1' and finit_state.isstate('s2'):
    finit_state.trigger (str_name)
    print('State',finit_state.current)
  elif str_name == '1' and finit_state.isstate('s3'):
    finit_state.trigger (str_name)
    print('State',finit_state.current)
  elif str_name == '1' and finit_state.isstate('s4'):
    finit_state.trigger (str_name)
    print('State',finit_state.current)
  else:
    print('ERROR')
    break
```
