```python
from fysom import Fysom
BinaryNumberSystem = Fysom({'initial' : 's0', 'events' : [
      {'name': '00', 'src': 's0', 'dst': 's0'},
      {'name': '00', 'src': 's1', 'dst': 's0'},
      {'name': '10', 'src': 's0', 'dst': 's0'},
      {'name': '10', 'src': 's1', 'dst': 's1'},
      {'name': '01', 'src': 's0', 'dst': 's0'},
      {'name': '01', 'src': 's1', 'dst': 's1'},
      {'name': '11', 'src': 's0', 'dst': 's1'},
      {'name': '11', 'src': 's1', 'dst': 's1'} ]})

number = int(input('number of Input ='))
print('Start state', BinaryNumberSystem.current)
for i in  range(number):
  name = input('Enter your input :')
  str_name = str(name)
  if str_name == '00'and BinaryNumberSystem.isstate('s0'):
    BinaryNumberSystem.trigger(str_name)
    print('State', BinaryNumberSystem.current)
    output = '0'
    print('output', output)
  elif str_name == '10'and BinaryNumberSystem.isstate('s0'):
    BinaryNumberSystem.trigger(str_name)
    print('State', BinaryNumberSystem.current)
    output = '1'
    print('output', output)
  elif str_name == '01'and BinaryNumberSystem.isstate('s0'):
    BinaryNumberSystem.trigger(str_name)
    print('State', BinaryNumberSystem.current)
    output = '1'
    print('output', output)
  elif str_name == '11'and BinaryNumberSystem.isstate('s0'):
    BinaryNumberSystem.trigger(str_name)
    print('State', BinaryNumberSystem.current)
    output = '0'
    print('output', output)
  elif str_name == '00'and BinaryNumberSystem.isstate('s1'):
    BinaryNumberSystem.trigger(str_name)
    print('State', BinaryNumberSystem.current)
    output = '1'
    print('output', output)
  elif str_name == '10'and BinaryNumberSystem.isstate('s1'):
    BinaryNumberSystem.trigger(str_name)
    print('State', BinaryNumberSystem.current)
    output = '0'
    print('output', output)
  elif str_name == '01'and BinaryNumberSystem.isstate('s1'):
    BinaryNumberSystem.trigger(str_name)
    print('State', BinaryNumberSystem.current)
    output = '0'
    print('output', output)
  elif str_name == '11'and BinaryNumberSystem.isstate('s1'):
    BinaryNumberSystem.trigger(str_name)
    print('State', BinaryNumberSystem.current)
    output = '1'
    print('output', output)
  else:
    print('ERROR')
    break
```

```python
def cross(A,B) :
  return {a+b for a in A for b in B}
A=set(['0','11'])
B=set(['1','10','110'])
AB = cross(A,B)
print("AB=",AB)
BA= cross(B,A)
print("BA=",BA)
print("AB = BA is ", AB == BA)
```

```python
from fysom import Fysom
finit_state = Fysom({ 'initial' : 's0','final':'s2', 'events' : [
      {'name': '0', 'src': 's0', 'dst': 's1'},
      {'name': '0', 'src': 's1', 'dst': 's3'},
      {'name': '0', 'src': 's2', 'dst': 's3'},
      {'name': '0', 'src': 's3', 'dst': 's3'},
      {'name': '1', 'src': 's0', 'dst': 's2'},
      {'name': '1', 'src': 's1', 'dst': 's2'},
      {'name': '1', 'src': 's2', 'dst': 's3'},
      {'name': '1', 'src': 's3', 'dst': 's3'} ]})

number = int(input('number of Input ='))
print('Start state',finit_state.current)
for i in range(number) :
  name = input('Enter your input :')
  str_name = str(name)
  if str_name == '0' and finit_state.isstate('s0'):
    finit_state.trigger(str_name)
    print('State',finit_state.current)
    print('Finished=',finit_state.is_finished())
  elif str_name == '0' and finit_state.isstate('s1'):
    finit_state.trigger(str_name)
    print('State', finit_state.current)
    print('Finished=', finit_state.is_finished())
  elif str_name == '0' and finit_state.isstate('s2'):
    finit_state.trigger(str_name)
    print('State', finit_state.current)
    print('Finished=', finit_state.is_finished())
  elif str_name == '0' and finit_state.isstate('s3'):
    finit_state.trigger(str_name)
    print('State', finit_state.current)
    print('Finished=', finit_state.is_finished())
  elif str_name == '1' and finit_state.isstate('s0'):
    finit_state.trigger(str_name)
    print('State', finit_state.current)
    print('Finished=', finit_state.is_finished())
  elif str_name == '1' and finit_state.isstate('s1'):
    finit_state.trigger(str_name)
    print('State',finit_state.current)
    print('Finished=', finit_state.is_finished())
  elif str_name == '1' and finit_state.isstate('s2'):
    finit_state.trigger(str_name)
    print('State', finit_state.current)
    print('Finished=', finit_state.is_finished())
  elif str_name == '1' and finit_state.isstate('s3'):
    finit_state.trigger(str_name)
    print('State', finit_state.current)
    print('Finished=', finit_state.is_finished())
  else:
    print('ERRO0R')
    break
```
